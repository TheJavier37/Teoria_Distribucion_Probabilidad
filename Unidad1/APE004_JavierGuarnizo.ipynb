```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# --- Configuración de Parámetros ---
nombre_estudiante = "Javier Guarnizo"
p = 0.5
dist_bernoulli = bernoulli(p)

# Definición del soporte (0=Cruz, 1=Cara)
x = [0, 1]
pmf_values = dist_bernoulli.pmf(x)
x_cdf = np.linspace(-0.5, 1.5, 1000)
cdf_values = dist_bernoulli.cdf(x_cdf)

# --- Visualización Profesional ---
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle(f'Análisis de Distribución Bernoulli - Estudiante: {nombre_estudiante}', fontsize=16)

# Gráfico 1: PMF
ax[0].stem(x, pmf_values, linefmt='navy', markerfmt='D', basefmt='gray')
ax[0].set_title('PMF: Probabilidad por Resultado', fontsize=12)
ax[0].set_xticks([0, 1])
ax[0].set_xticklabels(['0 (Cruz)', '1 (Cara)'])
ax[0].set_ylim(0, 1.1)
ax[0].set_ylabel('Probabilidad P(X=x)')
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 2: CDF
ax[1].plot(x_cdf, cdf_values, color='crimson', drawstyle='steps-post', linewidth=2.5)
ax[1].fill_between(x_cdf, cdf_values, step="post", alpha=0.1, color='crimson')
ax[1].set_title('CDF: Probabilidad Acumulada', fontsize=12)
ax[1].set_xticks([0, 1])
ax[1].set_ylabel('F(x) = P(X ≤ x)')
ax[1].grid(linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Cálculos Estadísticos ---
esperanza = dist_bernoulli.mean()
varianza = dist_bernoulli.var()

print(f"Reporte de {nombre_estudiante}")
print("-" * 50)
print(f"1. Probabilidad Exacta P(X=1) [Cara]: {dist_bernoulli.pmf(1):.4f}")
print(f"2. Probabilidad Acumulada P(X≤0) [Cruz]: {dist_bernoulli.cdf(0):.4f}")
print(f"3. Esperanza Matemática E[X]: {esperanza:.4f}")
print(f"4. Varianza Var(X): {varianza:.4f}")
print("-" * 50)

# --- Interpretación de Resultados ---
print("\nINTERPRETACIÓN DE LOS RESULTADOS:")
print(f"• La PMF muestra que existe un {dist_bernoulli.pmf(1)*100}% de probabilidad de éxito (Cara).")
print(f"• El valor esperado (E[X] = {esperanza}) indica que, en promedio, el resultado tiende al centro.")
print(f"• La varianza de {varianza} es la máxima posible para una Bernoulli (p=0.5), lo que")
print("  significa que este es el escenario de mayor incertidumbre: es tan probable ganar como perder.")
print(f"• La CDF confirma que la probabilidad acumulada llega a 1.0 tan pronto alcanzamos el valor '1'.")
```
